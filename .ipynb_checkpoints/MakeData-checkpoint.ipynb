```python
import pandas as pd
```

```python
#function for oxy sales
def make_sales_oxy():
    '''function to make df for sales data'''
    sales = pd.read_csv("https://raw.githubusercontent.com/VallieTracy/opioids_project/master/Resources/opioidPrescriptionSales.csv")
    sales['Type'] = sales['Oxycodone / Hydrocodone'].str.replace('/', '')
    sales = sales[["Fips", "Location","TimeFrame","Type", "Data", ]]
    sales=sales.query("Type == 'Oxycodone'")
    sales = sales.rename(columns={"Data":"OxycodoneSalesData"})
    return sales
make_sales_oxy()

# function for oxy sales
def make_sales_hydro():
    '''function to make df for sales data'''
    sales = pd.read_csv("https://raw.githubusercontent.com/VallieTracy/opioids_project/master/Resources/opioidPrescriptionSales.csv")
    sales['Type'] = sales['Oxycodone / Hydrocodone'].str.replace('/', '')
    sales = sales[["Fips", "Location","TimeFrame","Type", "Data", ]]
    sales= sales.query("Type == 'Hydrocodone'")
    sales=sales.query("Type == 'Hydrocodone'")
    sales = sales.rename(columns={"Data":"HydrocodoneSalesData"})
    return sales
make_sales_hydro()

# function to combine sales
def combo_sales():
    df1 = make_sales_oxy()
    df2 = make_sales_hydro()
    merged = df1.merge(df2,left_on=['Fips','Location','TimeFrame'], right_on=['Fips','Location','TimeFrame'])
    merged = merged[['Fips','Location','TimeFrame','OxycodoneSalesData','HydrocodoneSalesData']]
    return merged
combo_sales()

# function for deaths
def make_deaths():
    '''function to make df for deaths data'''
    deaths = pd.read_csv("https://raw.githubusercontent.com/VallieTracy/opioids_project/master/Resources/drugPoisoningDeaths.csv")
    deaths = deaths[["Fips", "Location","TimeFrame","Drug Type", "Data", ]]
    deaths= deaths.query("`Drug Type` == 'All opioids'")
    deaths = deaths.rename(columns={"Data":"DeathsData"})
    deaths = deaths.rename(columns={"Drug Type":"ODType"})
    return deaths
make_deaths()


# function to combine all data
def make_merge():
    '''function to combine death and sales data'''
    sales = combo_sales()
    deaths = make_deaths()
    #merged = deaths.merge(sales, on="Fips")
    merged= pd.merge(deaths, sales, left_on=['Fips','Location', 'TimeFrame'], right_on=['Fips','Location', 'TimeFrame'])
    merged = merged[['Fips','Location','TimeFrame','OxycodoneSalesData','HydrocodoneSalesData','DeathsData']]
    merged = merged.dropna()

    return merged
make_merge()
```

```python
#write to file
df = make_merge()
df.to_csv('data/SalesDeaths.csv')
```
